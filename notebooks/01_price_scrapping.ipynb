```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager as CM
from datetime import datetime, timedelta
import time
import sys
import warnings
import configparser
warnings.filterwarnings("ignore")

import sqlite3 as db
import numpy as np
import pandas as pd
```

```python
def get_login_from_config(filename='login.txt'):
    config = configparser.ConfigParser()
    print("[OK] Reading default config file...")
    try:
        config.read(filename)
    except:
        print(f"[ERROR] Config file not found. Please check {filename} in your current directory.")
    
    try:
        username = config.get("login", "mail")
        password = config.get("login", "password")
        print("[OK] Config file read successfully.")
    except:
        print(f"[ERROR] Email and password were not found / are not readable. Check {filename} in your current directory.")
        sys.exit()
    return username, password

def update_config(filename='login.txt'):
    # Ask the user for email and password
    username = input("Enter your mail: ")
    password = input("Enter your password: ")

    # the user can also save the new credentials to the config file (overwrite the old one)
    remember = input("Do you want to remember your password? (y/n) ")
    if remember == "y":
        config = configparser.ConfigParser()
        config.read(filename)
        config.set("login", "mail", username)
        config.set("login", "password", password)
        with open(filename, "w") as configfile:
            config.write(configfile)
    return username, password

# TO-DO: Make only valid inputs 1-0 with a while loop

def config(filename='login.txt'):
    # let the user choose if they want to use the config file or not
    choice = input("Do you want to use the default config file? (y/n) ")

    # if the user wants to use the config file, read it
    if choice == "y":
        username, password = get_login_from_config(filename)

    # if the user doesn't want to use the config file, ask for the credentials
    elif choice == "n":
        username, password = update_config(filename)
    else:
        print("Invalid input")
        sys.exit()
        
    return username, password

def create_driver(headless=True, incognito=True):
    # Set browser options
    options = webdriver.ChromeOptions()
    options.add_argument('--no-sandbox')
    options.add_argument("--log-level=3")
    options.add_argument("--silent")
    # Makes the browser run in background
    if headless:
        options.add_argument("--headless")
    # Makes the browser run in incognito mode
    if incognito:
        options.add_argument("--incognito")
    # Generate the driver
    browser = webdriver.Chrome(executable_path=CM().install(), options=options)
    print('[OK] Browser created')
    time.sleep(1)
    # Set window size to full screen
    browser.set_window_size(1920, 1080)
    print('[OK] Window size set to 1920x1080')
    time.sleep(2)
    return browser

def open_page(url='https://www.laligafantasymarca.com/'):
    # Open the Website
    try:
        browser.get(url)
        print('[OK] Page opened succesfully')
    except:
        print('[ERROR] Page cannot be opened. Check your internet connection')

def reject_cookies():
    print('[OK] [STEP 1] Rejecting Cookies...')
    time.sleep(3.5)
    # Access the login page step 1
    try:
        browser.find_element(By.CLASS_NAME,"cookie-setting-link").click()
        print('[OK] [STEP 1] Accesing Manage Preferences page')
    except:
        print('[ERROR] [STEP 1] Manage Preferences page cannot be accessed. Check your internet connection')
        
    time.sleep(1)
    try:
        browser.find_element(By.CLASS_NAME,"ot-pc-refuse-all-handler").click()
        print('[OK] [STEP 1] Refusing all cookies')
    except:
        print('[ERROR] [STEP 1] Something happend while refusing cookies. Check your internet connection')
        
        
def login_2():
    print('[OK] [STEP 2] Accessing the page...')
    time.sleep(2)
    # Access the login page step 2
    try:
        browser.find_element(By.CLASS_NAME,"btn-fantasy.grey.big.btn-secondary").click()
        print('[OK] [STEP 2] Login page accessed succesfully')
    except:
        print('[ERROR] [STEP 2] Login page cannot be accessed. Check your internet connection')


def input_credentials(email, password):
    time.sleep(2)
    # Fill credentials
    browser.find_element(By.NAME, 'username').send_keys(username)
    print('[OK] [STEP 3] Email entered succesfully')
    time.sleep(1)
    browser.find_element(By.NAME, 'password_current').send_keys(password)
    print('[OK] [STEP 3] Password entered succesfully')
    time.sleep(1.5)
    # Click Log In
    print('[OK] [STEP 3] Clicking Log In...')
    try:
        browser.find_element(By.CLASS_NAME,'btn-fantasy.green.big').click()
        time.sleep(3.5)
    except:
        print('[ERROR] [STEP 3] Log In button cannot be clicked. Check your internet connection')

def access_league():
    time.sleep(7)
    print('[OK] [STEP 4] Accessing the league...')
    # Accessing the league page
    try:
        browser.find_element(By.CLASS_NAME,"ltr.team").click()
        print('[OK] [STEP 4] League page accessed succesfully')
    except:
        print('[ERROR] [STEP 4] League page cannot be accessed. Check your internet connection')
        
        
def access_market():
    time.sleep(7)
    print('[OK] [STEP 5] Accessing the Market to get the info...')
    try:
        browser.find_element(By.XPATH,"/html/body/fy-app/fy-layout-game/div/div/div/fy-sidenav/aside/nav/div/ul/li[3]").click()
        print('[OK] [STEP 5] Market page accessed succesfully')
    except:
        print('[ERROR] [STEP 5] Market page cannot be accessed. Check your internet connection')
    # Switching to the search bar 
    time.sleep(1)
    try:
        browser.find_element(By.ID,"searchPlayer-tab").click()
        print('[OK] [STEP 5] succesfully accessed to search menu')
    except:
        print('[ERROR] [STEP 5] Market page cannot be accessed. Check your internet connection')
        
def show_all_players():
    time.sleep(1)
    print('[OK] [STEP 5] Showing All Players')
    try:
        browser.find_element(By.XPATH,"/html/body/fy-app/fy-layout-game/div/div/div/main/fy-league-market/div/div/fy-tabs-container/div/div[4]/div/div[1]/div/fy-search-player/div/div[4]/div/div/button").click()
        print('[OK] [STEP 5] Open dropdown')
    except:
        print('[ERROR] [STEP 5] Something went wrong openning dropdown')
    try:
        browser.find_element(By.XPATH,"/html/body/fy-app/fy-layout-game/div/div/div/main/fy-league-market/div/div/fy-tabs-container/div/div[4]/div/div[1]/div/fy-search-player/div/div[4]/div/div/div/a[1]").click()
        print('[OK] [STEP 5] All players now aviable')
    except:
        print('[ERROR] [STEP 5] Something went wrong while selectimg in dropdown')


def get_players_df():
    players = []
    for player in browser.find_elements(By.CLASS_NAME,"player-item"):
        text = player.text.replace('PFSY','\n').split(sep='\n')
        text  = list(map(lambda x: x.strip(), text))
        players.append(text)
    df = pd.DataFrame(players, columns= ['Name','Position','Owner','Price', 'Team', 'PSFY_current', 'PSFY_last_year'])
    df['Date_actua'] = pd.to_datetime("today").strftime("%d/%m/%Y")
    df['Price'] = df['Price'].str.replace('.','').astype('int')
    return df

def save_in_db(df): 
    conn = db.connect('../data/daily_price.db')
    df.to_sql("players_price", conn, if_exists="append")
    conn.close()
```
